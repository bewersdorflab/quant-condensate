```python
import tables
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import logging

logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)

# input glob ignored if recipe is already run
input_glob = '/media/smeagol/d5b5c790-4eba-4212-b990-062bba0b7a3b/data/cry2_cajal_bodies/STED_CBs/*.tif'
# results directory will be made by the program and filled. If it already exists, we assume the analysis recipe has already run
result_directory = '/media/smeagol/SSD/sted_cb_offset'
run_recipe = not os.path.exists(result_directory)
if run_recipe:
    print('no results found, will generate them in next cell')
```

```python
if run_recipe:  # PYME is only required if we need to run the recipe, import here
    from PYME.recipes.base import ModuleCollection
    from PYME.recipes import batchProcess
    import quant_condensate
    
    recipe_path = os.path.join(os.path.dirname(os.path.abspath(quant_condensate.__file__)),
                               'recipes', 'sted_measure.yaml')
    with open(recipe_path) as f:
        contents = f.read()
    recipe = ModuleCollection().fromYAML(contents)
    
    try:
        os.mkdir(result_directory)
    except:
        raise RuntimeError('Output directory already exists; exiting to avoid accidental replacement/appending to outputs')

    logger.addHandler(logging.FileHandler(os.path.join(result_directory, 'logging.txt')))
    logger.debug('initializing file logging')
    
    input_globs = {
        'input': glob.glob(input_glob)
    }
    print('%d inputs found' % len(input_globs['input']))
    batchProcess.bake(recipe, input_globs, result_directory, num_procs=1)
```

```python
# load and analyze results
files = glob.glob(os.path.join(result_directory, '*.hdf'))
d = dict()

for f in files:
    vals = os.path.split(os.path.splitext(f)[0])[-1].split('_')
    treat = vals[2]
    
    if treat not in d.keys():
        d[treat] = dict(path=[], centroid_offset=[])
    
    d[treat]['path'].append(f)
    
    t = tables.open_file(f)
    m0 = t.root.measurements0[:]
    m1 = t.root.measurements1[:]
    
    # remove measurements from labeled regions touching the edge of the field of view
    m0_edge = np.logical_or(np.any(m0['bbox'] == 0, axis=1), np.any(m0['bbox'] == 256, axis=1))
    if any(m0_edge):
        logger.error('removing edge label chan 0 from f:%s' % f)
        m0 = m0[~m0_edge]
    m1_edge = np.logical_or(np.any(m1['bbox'] == 0, axis=1), np.any(m1['bbox'] == 256, axis=1))
    if any(m1_edge):
        logger.error('removing edge label chan 1 from f:%s' % f)
        m1 = m1[~m1_edge]
    
    vx_um = t.get_node_attr(t.root.MetaData.voxelsize, 'x')
    vy_um = t.get_node_attr(t.root.MetaData.voxelsize, 'y')
    centroid_offset = np.sqrt(((vx_um * (m0['weighted_centroid'][:, 0] - m1['weighted_centroid'][:, 0])) ** 2)
                              + ((vy_um * (m0['weighted_centroid'][:, 1] - m1['weighted_centroid'][:, 1])) ** 2))
    
    if len(m0) != 1 or len(m1) != 1:
        logger.error('check segmentation/labels, chan 0 %d, chan 1 %d, f:%s' % (len(m0), len(m1), f))
    
    d[treat]['centroid_offset'].append(centroid_offset[0])
    if not (('SMN' in vals[0] and 'Coilin' in vals[1]) or ('SMN' in vals[1] and 'Coilin' in vals[0])):
        raise ValueError('unknown labeling')
    
    t.close()
```

```python
off = {
    'noTreat': d['noTreat']['centroid_offset'],
    'MS023': d['MS023']['centroid_offset']
}

plt.figure()
sns.boxplot(data=[off['noTreat'], off['MS023']])
sns.swarmplot(data=[off['noTreat'], off['MS023']], color=".2")
plt.xticks(plt.xticks()[0], ['noTreat', 'MS023'])#, rotation=45)
plt.ylabel('Offset [um]', fontsize=18)
```
